```python
import pandas as pd
```

```python
df_store = pd.read_csv("/content/stores_data_set.csv")
```

```python
df_sales = pd.read_csv("/content/sales_data_set.csv")
```

```python
df_feature = pd.read_csv("/content/Features_data_set.csv")
```

```python
df_store_sales = pd.merge(df_sales, df_store, on="Store")
```

```python
df_final=pd.merge(df_store_sales,df_feature,on=["Store","Date"])
```

```python
#1.)Identify average customer visit in Type C store in April Month.
```

```python
df_final
```

```python
df_final.drop("IsHoliday_y",axis=1,inplace=True)
```

```python
df_final["Date"] = pd.to_datetime(df_final["Date"],format="%d/%m/%Y")
```

```python
df_final["Month"] = df_final["Date"].dt.month
```

```python
df_final
```

```python
pd.set_option('display.max_rows', 10)
```

```python
((df_final[(df_final["Month"]==4)&(df_final["Type"]=="C")]\
            .groupby(["Store","Date"])["Weekly_Sales"].sum())\
           / (df_final[(df_final["Month"]==4)&(df_final["Type"]=="C")]\
              .groupby(["Store","Date"])["CPI"].unique())).groupby("Store").mean()
```

```python
((df_final[(df_final["Month"]==4)&(df_final["Type"]=="C")]\
.groupby(["Store","Date"])["Weekly_Sales"].sum())\
/ (df_final[(df_final["Month"]==4)&(df_final["Type"]=="C")]\
  .groupby(["Store","Date"])["CPI"].unique())).groupby("Store").mean()
```

```python
#2.)Find the highest sales in each store type on Holiday season.
df_final[(df_final["IsHoliday_x"]==True)].groupby("Type")["Weekly_Sales"].max()
```

```python
#3.)What is the expected sales of each depeartment when the unemployment factor is greater than 7.
```

```python
(df_final[(df_final["Unemployment"]>7)]["Weekly_Sales"].sum())/(df_final[(df_final["Unemployment"]>7)]["Weekly_Sales"].count())
```

```python
#4.)Aggregate net sales for each department in month wise.
df_final[:].groupby("Month")["Weekly_Sales"].sum()
```
